# Step 1: checking whether data is balanced or not

**Here the target variable is the STATUS of job**
- Now we have to check whether the target data is **balanced** or not 

```python
data['Status'].value_counts()/len(data)*100
# here we can see the percentage of each class 
```

- we can observe there is an imbalance in the data as one class is having percentage less than 20
- we can balance it by performing sampling
- The sampling method used here is **Upsampling**

# Step 2: Upsampling 

**Checking any missing data if present drop those data if it is less than 5%**

```python
data.isnull().sum()
```

```python
y=data.iloc[:,-1]
y
```

**replacing the categorical data with binary values** 

```python
y.replace(['Joined','Not Joined'],[0,1],inplace=True)#replace categorical data
```

**Getting the replaced values into the dataset**

```python
data['Status']=y
```

**Now again splitting the classes inside the data for sampling**

```python
#split the data into two parts
#sample=0 and sample=1
df_joined=data[data['Status']==0]
df_not_joined=data[data['Status']==1]
#to reduce biasing,the technique used is upsampling
```

**Importing the sampling library** 

```python
from sklearn.utils import resample
```

```python
df_no_sample=resample(df_not_joined,replace=True,n_samples=2000)
```

**Resampling the data which is low ,replacing it  and checking the % of the data**

```python
df_no_sample['Status'].value_counts()/len(data)*100
```

**Concatinating the sampled data into the main data**

```python
data_resam=pd.concat((df_not_joined,df_joined,df_no_sample),axis=0)
data_resam
```

```python
data_resam['Status'].value_counts()
```

**Here we got the balanced data and saving it for further use**


```python
data_resam.to_csv('data_HR_resample.csv')
data_resam.head()
```

# Step 6: Maximum Likelihood Estimation(MLE)

```python
from sklearn.metrics import log_loss
```

```python
class threshold():
    def __init(self):
        self.th=0.5
    def predict_threshold(self,y):
        if y>=self.th:
            return 1
        else:
            return 0
```

```python
y_pred_prob_1=y_pred_train_1
probabilities=np.linspace(0,1,num=100)#generating 100 values in 0 to 1
pre=threshold()
mle=[]
for p in probabilities:
    pre.th=p
    yy=pd.Series(y_pred_prob_0)
    y_pred_1_=yy.apply(pre.predict_threshold)
    loss=log_loss(y_train,y_pred_1_)
    mle.append(loss)
```

**Here predicting the likelihood of candidate not joining using the log loss function**

```python
plt.figure(figsize=(30,10))
plt.plot(probabilities,mle,'ro--')
plt.plot(probabilities,mle,'b--')
ind=np.array(mle).argmin()
mx=max(mle)
val=probabilities[ind]
print('Threshold probability value',val)
plt.plot([val,val],[0,mx],'g')
for i,p in enumerate(probabilities):
    plt.text(probabilities[i],mle[i],'%0.2f'%(p))

    
    
plt.xlabel('Probabilities')
plt.ylabel('-(Log Likelihood)')
plt.title('Maximum Likelihood Estimation')
plt.show()
```

# Oversampling

- SMOTE(Synthetic Minority Oversampling TechniquE)

```python
import imblearn
from imblearn.over_sampling import SMOTE
```

**Importing the neccessary libraies and resampling the data using SMOTE**

```python
x_resample,y_resample=SMOTE(k_neighbors=5).fit_resample(Xnew3,Y)
```

```python
#saving array's
np.savez('hr_oversample.npz',x_resample,y_resample)
```

```python
pd.Series(y_resample).value_counts().plot('bar')
pd.Series(y_resample).value_counts()
```

**We can see the equally sampled data of both classes**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_resample,y_resample,test_size=0.2)#30% data to test
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_over=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
```

```python
model_over.fit(x_train,y_train)
```

**Modelling the data after oversampling**

```python
y_pred=model_over.predict(x_test)
y_pred_prob=model_over.predict_proba(x_test)[:,1]#considering 1 probability
```

**Predicting the probabilities and calculating the metrics**

```python
innomatics.classification_metric(y_test,y_pred,y_pred_prob,['no','yes'],n=1,verbose=True)
```

**I have got slightly good kappa score**

**Criteria is to have the """Cost of predicting â€œNot Joiningâ€ as â€œJoiningâ€ (FPs) cases is 3 times more than predicting â€œJoiningâ€ as â€œNot Joiningâ€ (FNs)"""

**Here setting the threshold value obtained from the ROC curve in order to get required confusion matrix**

```python
y_pred_prob_test=model.predict_proba(x_test)
y_pred_prob_test_1=y_pred_prob_test[:,1]
```

```python
pre=innomatics.threshold()#predictor class
threh_prob=0.2849# from ROC curve
pre.th=threh_prob#setting threshold
yy=pd.Series(y_pred_prob_test_1)#converting prob array into series
y_pred_class_test=yy.apply(pre.predict_threshold)#apply predictor class
```

```python
cm=confusion_matrix(y_test,y_pred_class_test)
cm
```

**The criteria is satisfied by setting the threshold value**
